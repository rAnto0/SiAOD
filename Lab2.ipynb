```python
from collections import deque
from pathlib import Path

DATA_DIR = Path("Lab2")
```

### Задание 1


```python
def sort_book_titles():
    with open(DATA_DIR / 'book.txt', 'r', encoding='utf-8') as file:
        lines = [line.strip() for line in file]

    original_deque = deque(sorted(lines))
    sorted_deque = deque()

    while original_deque:
        sorted_deque.append(original_deque.pop())

    with open(DATA_DIR / 'sorted_book.txt', 'w', encoding='utf-8') as file:
        while sorted_deque:
            file.write(sorted_deque.pop() + '\n')

sort_book_titles()
```

### Задание 2


```python
def deshifr(ciphertext, deck):
    decrypted_text = ''

    for char in ciphertext:
        if char.isalpha():
            idx = deck.index(char)
            decrypted_text += deck[(idx - 2) % 26]
        else:
            decrypted_text += char

    return decrypted_text


def shifr(text, deck):
    encrypted_text = ''

    for char in text:
        if char.isalpha():
            idx = deck.index(char)
            encrypted_text += deck[(idx + 2) % 26]
        else:
            encrypted_text += char

    return encrypted_text


deck = deque('ABCDEFGHIJKLMNOPQRSTUVWXYZ')

with open(DATA_DIR / 'зашифрованное.txt', 'r', encoding='utf-8') as file:
    with open(DATA_DIR / 'расшифрованное.txt', 'w', encoding='utf-8') as output_file:
        for line in file:
            decrypted_line = deshifr(line.strip().upper(), deck)
            output_file.write(decrypted_line + '\n')
```

### Задание 3


```python
def tower_of_hanoi(n, source, destination, auxiliary):
    if n == 1:
        print(f"Move disk 1 from peg source({source}) to peg destination({destination})")
        destination.append(source.pop())
        print(source, destination, auxiliary)
        return

    tower_of_hanoi(n - 1, source, auxiliary, destination)
    print(f"Move disk {n} from peg source({source}) to peg destination({destination})")
    destination.append(source.pop())
    print(source, destination, auxiliary)
    tower_of_hanoi(n - 1, auxiliary, destination, source)

with open(DATA_DIR / 'данные_о_дисках.txt', 'r', encoding='utf-8') as file:
    disks = [int(line.strip()) for line in file]

stack_a = list(disks)
stack_b = []
stack_c = []

print('Стек A:', stack_a)
print('Стек B:', stack_b)
print('Стек C:', stack_c)

tower_of_hanoi(len(disks), stack_a, stack_c, stack_b)

print('Стек A:', stack_a)
print('Стек B:', stack_b)
print('Стек C:', stack_c)
```

### Задание 4


```python
def check_bracket_balance():
    stack = []
    open_brackets = {'(', '{', '['}
    close_brackets = {')': '(', '}': '{', ']': '['}

    with open(DATA_DIR / 'program_file.txt', 'r', encoding='utf-8') as file:
        for line_num, line in enumerate(file, start=1):
            for char_num, char in enumerate(line, start=1):
                if char in open_brackets:
                    stack.append((char, line_num, char_num))
                elif char in close_brackets:
                    if not stack or stack[-1][0] != close_brackets[char]:
                        print(f"Ошибка: Лишняя закрывающая скобка '{char}' в строке {line_num}, позиция {char_num}")
                        return
                    stack.pop()

    for bracket, line_num, char_num in stack:
        print(f"Ошибка: Лишняя открывающая скобка '{bracket}' в строке {line_num}, позиция {char_num}")

    if not stack:
        print('Баланс скобок соблюден.')
    else:
        print('Ошибка: Остались несогласованные открывающие скобки.')

check_bracket_balance()
```

### Задание 5


```python
def check_square_bracket_balance():
    stack = deque()

    with open(DATA_DIR / 'program_file2.txt', 'r', encoding='utf-8') as file:
        for line_num, line in enumerate(file, start=1):
            for char_num, char in enumerate(line, start=1):
                if char == '[':
                    stack.append(('[', line_num, char_num))
                elif char == ']':
                    if not stack or stack[-1][0] != '[':
                        print(f"Ошибка: Лишняя закрывающая квадратная скобка ']' в строке {line_num}, позиция {char_num}")
                        return
                    stack.pop()

    for bracket, line_num, char_num in stack:
        print(f"Ошибка: Лишняя открывающая квадратная скобка '{bracket}' в строке {line_num}, позиция {char_num}")

    if not stack:
        print('Баланс квадратных скобок соблюден.')
    else:
        print('Ошибка: Остались лишние открывающие квадратные скобки.')

check_square_bracket_balance()
```

### Задание 6


```python
def print_characters_in_groups():
    digits_stack = []
    letters_stack = []
    other_stack = []

    with open(DATA_DIR / 'file_example6.txt', 'r', encoding='utf-8') as file:
        for line in file:
            for char in line:
                if char.isdigit():
                    digits_stack.append(char)
                elif char.isalpha():
                    letters_stack.append(char)
                else:
                    other_stack.append(char)

    print('Цифры:', *digits_stack)
    print('Буквы:', *letters_stack)
    print('Остальные символы:', other_stack)

print_characters_in_groups()
```

### Задание 7


```python
def print_numbers_in_groups():
    negative_numbers = deque()
    positive_numbers = deque()

    with open(DATA_DIR / 'example_numbers7.txt', 'r', encoding='utf-8') as file:
        for line in file:
            numbers = [int(num) for num in line.split()]
            for num in numbers:
                if num < 0:
                    negative_numbers.append(num)
                else:
                    positive_numbers.append(num)

    print('Отрицательные числа:', end=' ')
    while negative_numbers:
        print(negative_numbers.popleft(), end=' ')
    print()

    print('Положительные числа:', end=' ')
    while positive_numbers:
        print(positive_numbers.popleft(), end=' ')
    print()

print_numbers_in_groups()
```

### Задание 8


```python
def reverse_file_content():
    stack = []

    with open(DATA_DIR / 'file_example8.txt', 'r', encoding='utf-8') as file:
        for line in file:
            stack.append(line.strip())

    with open(DATA_DIR / 'output_file_example8.txt', 'w', encoding='utf-8') as output_file:
        while stack:
            output_file.write(stack.pop() + '\n')

reverse_file_content()
```
